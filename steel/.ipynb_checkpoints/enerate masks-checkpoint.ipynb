```python
#assuming masks dont overlap
def rle_decode(mask_rle,shape=(1400,2100)):
    '''
    mask_rle: run-length as string formated (start length)
    shape: (height,width) of array to return 
    Returns numpy array, 1 - mask, 0 - background
    '''
    s = mask_rle.split()
    starts, lengths = [np.asarray(x, dtype=int) for x in (s[0:][::2], s[1:][::2])]
    starts -= 1
    ends = starts + lengths
    img = np.zeros(shape[0] * shape[1], dtype=np.uint8)
    
    for lo, hi in zip(starts, ends):
        img[lo:hi] = 1
        
    return img.reshape(shape,order='F')
```

```python
def bin2array(num):
    if num == 1:
        return [1]
    elif num == 2:
        return [2]
    elif num == 3:
        return [1,2]
    elif num == 4:
        return [3]
    elif num == 5:
        return [1,3]
    elif num == 6:
        return [2,3]
    elif num == 7:
        return [1,2,3]
    elif num ==8:
        return [4]
    elif num ==9:
        return [1,4]
    elif num == 10:
        return [2,4]
    elif num == 11:
        return [1,2,4]
    elif num == 12:
        return [3,4]
    elif num ==13:
        return [1,3,4]
    elif num ==14:
        return [2,3,4]
    elif num == 15:
        return [1,2,3,4]
```

```python
img_sz = (256,1600)
```

```python
def img2mask(img_fn):
    msk = np.zeros(256*1600,dtype=np.uint8)
    #print("ravi:start \t",img_fn,fls2defects[img_fn])
    if fls2defects[img_fn] == 0:
        return msk.reshape(img_sz,order='F')
    else:
        for i in bin2array( fls2defects.get(img_fn) ):
            #print("i\t",i)
            string = df_enc[df_enc[df_enc.columns[0]] == (img_fn+"_"+str(i)) ][df_enc.columns[1]].values[0]
            #print("ravi:\t",string)
            msk = rle_decode(string,msk,i)
            #print(np.unique(msk))
        return msk.reshape(img_sz,order='F')
```
